### 1. 뉴스 기사 데이터 크롤링

1.1 원하는 뉴스 기사의 본문을 크롤링하시오.

```python
import pandas as pd
from newspaper import Article

url = 'https://www.sisajournal-e.com/news/articleView.html?idxno=401374'

article= Article(url, language='ko')
article.download()
article.parse()

print('기사 제목 :')
print(article.title)
print()

print('기사 내용 :')
content = article.text
print(content)
```

1.2 크롤링한 본문을 단어 단위로 나누고 각 단어가 나온 횟수와 함께 dataframe 형식으로 변환하시오.

```python
from collections import Counter

words = content.split()

# 각 단어의 빈도 계산
word_counts = Counter(words)

# DataFrame으로 변환
df = pd.DataFrame(word_counts.items(), columns=['단어', '빈도'])
df
```
