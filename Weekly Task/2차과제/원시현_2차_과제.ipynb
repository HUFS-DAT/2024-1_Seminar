1.3 불용어를 제거하시오.

```python
string = '\n'.join(str(s) for s in text_list) #단어 리스트를 문자열로 변환
print(string)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
from konlpy.tag import Okt
```

```python
okt = Okt() 
#불용어제거
stop = '해야 ( 60·朱锋 ) 16일 ( TCS ) 개최 그릇 했다 제1 들 없다 은 '

word_tokens = word_tokenize(string)
stop_words = stop.split(' ')

news_words = [] 
for w in word_tokens: 
    if w not in stop_words: 
        news_words.append(w) 

news_words
```

```python
news_df = pd.DataFrame({'word' : news_words})

news_df = news_df.groupby('word')\
                 .agg(n = ('word', 'count')) \
                 .sort_values('n', ascending = False)
news_df
```

```python
news_i = news_df.reset_index() #인덱스 추가
news_i
```

```python
news_i.rename(columns = {'word':'단어', 'n':'단어수'}, inplace=True)
news_i
```

```python
from selenium.webdriver.common.by import By

# 최하단으로 스크롤내리기
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
#리뷰모두보기클릭
spread_review = driver.find_element(by=By.XPATH, value = '//*[@id="yDmH0d"]/c-wiz[2]/div/div/div[1]/div/div[2]/div/div[1]/div[1]/c-wiz[5]/section/div/div[2]/div[5]/div/div/button/span')
isTrue = spread_review.is_displayed()
if isTrue :
    driver.execute_script("arguments[0].click();", spread_review)
    time.sleep(1.5)
```

```python
#리뷰창에서 스크롤내리기
all_reviews = driver.find_element(by=By.XPATH, value ='//*[@id="yDmH0d"]/div[4]/div[2]/div/div/div/div/div[2]')

for i in range(10):
    driver.execute_script('arguments[0].scrollTop = arguments[0].scrollHeight', all_reviews)
    time.sleep(3)
```

2.2 크롤링한 데이터를 dataframe 형태로 변환하시오(날짜, 별점, 리뷰 유저 등).

```python
import pandas as pd

data = pd.DataFrame(data=[], columns=['날짜','리뷰','별점'])
#날짜, 리뷰, 별점 수집
dates = driver.find_elements(by=By.XPATH, value = '//span[@class="bp9Aid"]')
reviews = driver.find_elements(by=By.XPATH, value = '//div[@class="h3YV2d"]')
stargrades = driver.find_elements(by=By.XPATH, value = '//div[@class="iXRFPc"]')

for i in range(len(reviews)):
    tmp = []
    tmp.append(dates[i].text)
    tmp.append(reviews[i].text)
    tmp.append(stargrades[i].get_attribute('aria-label'))

    tmp = pd.DataFrame(data=[tmp], columns = data.columns)
    data = pd.concat([data,tmp])
    
print("비리얼 앱 리뷰 수집 완료")


data.reset_index(inplace=True, drop=True)
data
```

2.3 긍정 리뷰와 부정 리뷰를 나누시오(기준은 자유).

```python
import re

tmp = data.copy()

tmp['별점'] = tmp['별점'].apply(lambda x : x[5:])

m = re.compile('[0-9][\.0-9]*') #정규표현식

tmp['별점'] = tmp['별점'].apply(lambda x : m.findall(x)[0])
tmp['별점']
```

```python
#별점 type변환
tmp.별점 = tmp.별점.astype(float)

#별점 나누기
tmp_low=tmp[tmp['별점']<=3]
tmp_high=tmp[tmp['별점']>3]
```
