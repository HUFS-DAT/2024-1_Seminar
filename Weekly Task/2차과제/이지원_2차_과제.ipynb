### 1. 뉴스 기사 데이터 크롤링

1.1 원하는 뉴스 기사의 본문을 크롤링하시오.

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://n.news.naver.com/mnews/article/016/0002282100"
response = requests.get(url)
response.raise_for_status()

web = requests.get(url).content
source = BeautifulSoup(web, 'html.parser')
```

```python
source.find('h2').get_text().strip()
```

```python
article = source.find('div', {'id': 'newsct_article'}).get_text().strip()
article
```

```python
# 필요없는 부분 삭제

del_index = article.find('기자]')
article[:del_index+4] 
article = article[del_index+4:]
article
```

1.2 크롤링한 본문을 단어 단위로 나누고 각 단어가 나온 횟수와 함께 dataframe 형식으로 변환하시오.

```python
article_list = article.split(" ")
article_list
```

```python
word_check = {}
for word in article_list:
    if word not in word_check:
        word_check[word] = 1
    else:
        word_check[word] += 1
```

```python
word = [*word_check.keys()]
word_count = [*word_check.values()]
word, word_count
```

```python
import pandas as pd
df_article = pd.DataFrame({'word':word, 'word_count':word_count})
df_article
```

1.3 불용어를 제거하시오.

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
```

```python
import nltk
nltk.download('punkt')
```

```python
# 제거할 불용어
stop = "이,있,하,것,들,그,되,수,이,보,않,지,한,(,),%,@,[,],.,‘,”,’,기자/20ki"
# 리스트를 스트링으로 변환
word_string = " ".join(df_article['word'])

stop_list = stop.split(',')
tok = word_tokenize(word_string)

def stopword(word_tokenize):
    result = []
    
    for w in word_tokenize:
        if w not in stop_list:
            result.append(w)
            
    return result

print('불용어를 제거하면' + str(stopword(tok)) + '입니다.')
```

```python
word_check = {}
for word in stopword(tok):
    if word not in word_check:
        word_check[word] = 1
    else:
        word_check[word] += 1

word = [*word_check.keys()]
word_count = [*word_check.values()]
word, word_count

df_new = pd.DataFrame({'word':word, 'word_count':word_count})
df_new
```

```python
# 카카오 앱 리뷰 url
url = 'https://play.google.com/store/apps/details?id=com.kakao.talk&hl=ko-KR'
driver.switch_to.window(driver.window_handles[0])

# 페이지 열기
driver.get(url)
# 페이지 로딩 대기
wait = WebDriverWait(driver, 5)
```

```python
from selenium.webdriver.common.by import By

#날짜, 리뷰, 별점 수집
dates = driver.find_elements(by=By.XPATH, value = '//span[@class="bp9Aid"]')
reviews=driver.find_elements(by=By.XPATH, value = '//div[@class="h3YV2d"]')
stargrades = driver.find_elements(by=By.XPATH, value = '//div[@class="iXRFPc"]')
```

```python
# 리뷰
reviews=driver.find_elements(by=By.XPATH, value = '//div[@class="h3YV2d"]')
for review in reviews:
    print(review.text)

print(len(reviews))
```

```python
# 최하단으로 스크롤내리기
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

#리뷰전체보기클릭
spread_review = driver.find_element(by=By.XPATH, value = '//*[@id="yDmH0d"]/c-wiz[2]/div/div/div[1]/div/div[2]/div/div[1]/div[1]/c-wiz[5]/section/div/div[2]/div[5]/div/div/button')
isTrue = spread_review.is_displayed()
if isTrue :
    driver.execute_script("arguments[0].click();", spread_review)
    time.sleep(1.5)
```

```python
# xpath를 복사하여 all_review로 해당 팝업을 지정해주고
# all review를 인자로 받아서 스크롤을 내리는 함수를 구현해주면된다.
# 기존에 무한스크롤함수로는 스크롤이 불가하여 for loop를 돌면서 스크롤 높이를 변경하는 식으로 구현하였다.

all_reviews = driver.find_element(by=By.XPATH, value ='//*[@id="yDmH0d"]/div[4]/div[2]/div/div/div/div')

for i in range(10):
    driver.execute_script('arguments[0].scrollTop = arguments[0].scrollHeight', all_reviews)
    time.sleep(3)
```

2.2 크롤링한 데이터를 dataframe 형태로 변환하시오(날짜, 별점, 리뷰 유저 등).

```python
data = pd.DataFrame(data=[], columns=['날짜','리뷰','별점', '리뷰 공감수'])
#날짜, 리뷰, 별점 수집
dates = driver.find_elements(by=By.XPATH, value = '//span[@class="bp9Aid"]')
reviews=driver.find_elements(by=By.XPATH, value = '//div[@class="h3YV2d"]')
stargrades = driver.find_elements(by=By.XPATH, value = '//div[@class="iXRFPc"]')
other_rate_reviews = driver.find_elements(by=By.XPATH, value = '//div[@class="AJTPZc"]')

# k개의 리뷰를 수집합니다.
for i in range(len(reviews)):
    tmp = []
    tmp.append(dates[i].text)
    tmp.append(reviews[i].text)
    tmp.append(stargrades[i].get_attribute('aria-label'))
    tmp.append(other_rate_reviews[i].text)

    tmp = pd.DataFrame(data=[tmp], columns = data.columns)
    data = pd.concat([data,tmp])
    
print("카카오톡 앱 리뷰 수집 완료")


data.reset_index(inplace=True, drop=True)
data
```

2.3 긍정 리뷰와 부정 리뷰를 나누시오(기준은 자유).

```python
# 별점에서 숫자만 가져오기

import re

tmp = data.copy()

tmp['별점'] = tmp['별점'].apply(lambda x : x[5:])

m = re.compile('[0-9][\.0-9]*') #정규표현식

tmp['별점'] = tmp['별점'].apply(lambda x : m.findall(x)[0])
tmp['별점']
```

```python
#별점 type변환
tmp.별점 = tmp.별점.astype(float)
#별점 나누기
tmp_low=tmp[tmp['별점']<=3]  # 부정 리뷰
tmp_high=tmp[tmp['별점']>3]  # 긍정 리뷰
```

```python
# 긍정 리뷰 data
tmp_high
```

```python
# 부정 리뷰 data
tmp_low
```

```python
tmp = tmp.iloc[1:,]
new = tmp.copy()

new['리뷰 공감수'] = new['리뷰 공감수'].apply(lambda x : x[4:8])

m = re.compile('[0-9][\.0-9]*') #정규표현식

new['리뷰 공감수'] = new['리뷰 공감수'].apply(lambda x : m.findall(x)[0])
new['리뷰 공감수'] = new['리뷰 공감수'].astype(int)
new
```
