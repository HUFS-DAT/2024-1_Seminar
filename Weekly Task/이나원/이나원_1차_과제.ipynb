### 2.6 1,2 -> 상위, 4,5-> 하위로 묶은 후 One Hot Encoding을 통한 dummy variable 변환

```python
def categorize(grade):
    if grade == 1.0 or grade == 2.0:
        return '상위'
    elif grade == 4.0 or grade == 5.0:
        return '하위'
    else:
        return '중위'
df['학생성적구분코드'] = df['학생성적구분코드'].apply(categorize)
df.groupby('학생성적구분코드').size()
```

```python
df = pd.get_dummies(df,columns=['학생성적구분코드'],dtype=int)
df.rename(columns={"학생성적구분코드_중위":"중위","학생성적구분코드_상위":"상위",'학생성적구분코드_하위':'하위'},inplace=True)
df
```

### 2.7 label encoding과 One Hot Encoding의 차이
#### < 범주형 변수 처리>
**label encoding**
- 각 변수별 속성에 알파벳 순서에 따라 unique한 정수 할당
    ex )india, us, japan 이 있다면 각각 0, 2, 1 
- 속성값을 정수로 바꿔주는 것이기 때문에 dataframe 자체의 크기 변화 없음
- 순서가 없는 칼럼에 수행하면 알파벳순으로 랭크가 되기 때문에 모델에 잘못 반영될 수 있다

**One-Hot Encoding**
- 범주형 변수를 0 또는 1 값을 가진 새로운 특성으로 바꾼 것 
- 칼럼을 목록화하여 변수의 차원을 늘린다

### 2.9 다중공선성 문제가 무엇인지 작성하고 왜 이것이 One Hot Encoding의 dummy variable과 관련 있을까? 
#### 다중공선성 문제 
- 독립변수들 간의 강한 상관관계를 나타나는 문제
- 진단
    - 결정계수가 높지만, 독립변수의 p-value가 커서 개별 인자가 유의하지 않는 경우 의심
    - 독립변수들관 상관계수
    - 분산팽창요인(VIF) 값이 10이 넘으면 보통 다중공선성 문제 존재
    
#### dummy variable과 다중공선성 문제
- One-Hot Encoding을 통해 모든 범주에 하나의 더미변수를 만들면 하나의 더미변수를 다른 더미변수의 선형 조합으로 예측이 가능하기 때문에 관련이 있다
- 해결하기 위해서 더미변수를 (범주의 개수 -1)개 생성한다
