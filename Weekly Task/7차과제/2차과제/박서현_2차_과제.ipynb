### 1. 뉴스 기사 데이터 크롤링

1.1 원하는 뉴스 기사의 본문을 크롤링하시오.

```python
import requests
from bs4 import BeautifulSoup

url='https://biz.heraldcorp.com/view.php?ud=20240319050187'

response = requests.get(url)
html = response.text
soup = BeautifulSoup(html, "html.parser")
articles = soup.select("div.info_group")

for article in articles:
    links = article.select("a.info")
    if len(links) >= 2:
        url = links[1].attrs["href"]
```

```python
import time
for article in articles:
    links = article.select("a.info")
    if len(links) >= 2:
        url = links[1].attrs["href"]
        response = requests.get(url, headers={'User-agent': 'Mozila/5.0'})      # to avoid error use headers
        html = response.text                                                    # for each url get html
        soup = BeautifulSoup(html, "html.parser")                               # for each html make soup
        content = soup.select_one("#dic_area")                                  # get the body
        print(content.text)
        time.sleep(0.3)
```

```python
title = soup.find('div', class_= 'article_top').text
print(title)
```

```python
article = soup.find('div',class_= 'article_view',itemprop="articleBody")
print(article)
```

```python
article_text = article.text.strip()
article_text
```

1.2 크롤링한 본문을 단어 단위로 나누고 각 단어가 나온 횟수와 함께 dataframe 형식으로 변환하시오.

```python
import nltk
from nltk.tokenize import word_tokenize
import collections
import pandas as pd

words = nltk.word_tokenize(article_text)
word_counts = collections.Counter(words)
df = pd.DataFrame(word_counts.items(), columns=['words', 'cnt'])
df = df.sort_values(by='cnt', ascending=False)
print(df)
```

1.3 불용어를 제거하시오.

```python
pip install konlpy
```

```python
from konlpy.corpus import kobill
from konlpy.tag import Okt
from nltk.corpus import stopwords

okt=Okt() # Okt 형태소 분석기 초기화
words = okt.nouns(article_text)
tokens = okt.morphs(article_text)

print('토큰:',tokens)
print('단어:',words)

# 불용어 제거
stop_words = set(['은', '는', '이', '가', '을', '를', '의', '에', '와', '과', '한', '하다','것','약','위해','할', '수', '하고','라고','로','으로','할','수','보','주','등','의','-','.','"','"','(',')']) # 한글 불용어 사전 로드
filtered_tokens = [word for word in tokens if word not in stop_words]
filtered_words = [word for word in words if word not in stop_words]

# 결과 출력
print("불용어 제거 토큰:", filtered_tokens)
print("불용어 제거 단어:", filtered_words)
```
