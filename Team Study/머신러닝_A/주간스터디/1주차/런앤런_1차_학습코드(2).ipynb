```python
# 빙어 도미 길이, 무게 데이터를 각각 따로 만들어준다.

fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8,
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]

fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7,
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
# fish_length와 fish_weight를 합쳐 데이터셋을 만들어주고 정답 데이터를 만들어준다.

fish_data = [[l, w] for l, w in zip(fish_length, fish_weight)]
fish_target = [1]*35 + [0]*14
```

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()
```

```python
# fish_data를 훈련 데이터와 테스트 데이터로 나눠준다.

# 훈련 세트 0 ~ 34 인덱스
train_input = fish_data[:35]

# 훈련 세트로 타깃값 중 0~34번째 인덱스 사용
train_target = fish_target[:35]

# 테스트 세트 35~
test_input = fish_data[35:]

# 테스트 세트로 타깃값 중 35~
test_target = fish_target[35:]
```

```python
# fit(훈련 데이터셋, 훈련 정답 데이터셋)
kn = kn.fit(train_input, train_target)
# score(테스트 데이터셋, 테스트 정답 데이터셋)
kn.score(test_input, test_target)
```

```python
import numpy as np

# 파이썬 리스트를 넘파이 배열로 바꾸기
input_arr = np.array(fish_data)
target_arr = np.array(fish_target)

print(input_arr)
```

```python
# 배열의 크기를 알려주는 shape 속성

print(input_arr.shape)
```

```python
# 인덱스 번호를 랜덤하게 섞는 방식으로 데이터를 섞어준다

# 데이터 섞기
# arrange() 함수를 통해 0~48까지 1씩 증가하는 인덱스를 만들 수 있음
# 그 다음 인덱스를 랜덤하게 섞음

np.random.seed(42) # 책과 동일한 실습 결과를 얻을 수 있도록 랜덤 시드 42 지정
index = np.arange(49) # 인덱스를 만든다
np.random.shuffle(index) # 인덱스를 랜덤하게 섞는다

print(index)
```

```python
# 넘파이는 배열 인덱싱을 통해 여러 개의 인덱스로 한 번에 여러 개의 원소 선택 가능

print(input_arr[[1, 3]])
```

```python
# 위 방법과 비슷하게 index 배열의 처음 35개를 전달하여 랜덤하게 훈련 세트 만들기

train_input = input_arr[index[:35]]
train_target = target_arr[index[:35]]

# index 첫 번째 값이 13이므로 train_input의 첫번째 원소는 input_arr의 14번째 원소가 들어감
print(input_arr[13], train_input[0])
```

```python
# 나머지 14개를 테스트 세트로

test_input = input_arr[index[35:]]
test_target = target_arr[index[35:]]
```

```python
import matplotlib.pyplot as plt

# 2차원 배열은 행과 열 인덱스를 콤마로 나누어 지정 가능
plt.scatter(train_input[:, 0], train_input[:, 1])
plt.scatter(test_input[:,0], test_input[:, 1])
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

# 파란색이 훈련 세트 / 주황색이 테스트 세트
```

```python
# 데이터를 섞어준 다음 모델링시도했다.
kn = kn.fit(train_input, train_target)

kn.score(test_input, test_target) # 100 % 정확도
```
