## 첫 번째 머신러닝 프로그램

```python
length = bream_length+smelt_length
weight = bream_weight+smelt_weight

fish_data = [[l,w] for l, w in zip(length,weight)]
print(fish_data)
```

```python
fish_target = [1]*35 + [0]*14
print(fish_target)
```

##K-최근접 이웃 알고리즘
fit메서드에 전달한 데이터 모두 저장, 새로운 데이터 등장하면 가장 가까운 데이터 참고하여 분류  

데이터가 아주 많은경우 사용하기 어려움
- 데이터가 크기에 메모리 많이 필요하고, 직선거리를 계산하는데도 많은 시간이 필요

```python
from sklearn.neighbors import KNeighborsClassifier

kn=KNeighborsClassifier()
kn.fit(fish_data, fish_target) #훈련
kn.score(fish_data, fish_target) #정확도
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.scatter(30, 600, marker='^') #길이 30 무게 600의 물고기 표시
plt.xlabel('lenght')
plt.ylabel('weight')
plt.show()
```

```python
kn.predict([[30,600]]) # 예측 , 이차원 배열로 넣어야함, 도미로 분류
```

```python
print(kn._fit_X)#_fit_X 속성에 전달한 데이터 모두를 저장하고 있음
```

```python
print(kn._y)
```

```python
kn49 = KNeighborsClassifier(n_neighbors=49) #기본은 5개의 데이터
#가장가까운 49개의 데이터로 변경

kn49.fit(fish_data, fish_target)
kn49.score(fish_data, fish_target)
```

```python
print(35/49)#가장 가까운 49개 중 35개가 도미
```

### 확인 문제

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5, 50):
    # 최근접 이웃 개수 설정
    kn.n_neighbors = n
    # 점수 계산
    score = kn.score(fish_data, fish_target)
    # 100% 정확도에 미치지 못하는 이웃 개수 출력
    if score < 1:
        print(n, score)
        break
```
