**2. 크롤링한 텍스트 데이터를 토큰화하고, 각 단어의 빈도를 계산하세요.**

```python
import nltk
from nltk.tokenize import word_tokenize
import collections
import pandas as pd

words = nltk.word_tokenize(article_text)
word_counts = collections.Counter(words)
df = pd.DataFrame(word_counts.items(), columns=['words', 'cnt'])
df = df.sort_values(by='cnt', ascending=False)
print(df)
```

```python
#불용어 제거
from konlpy.corpus import kobill
from konlpy.tag import Okt
from nltk.corpus import stopwords

okt=Okt() # Okt 형태소 분석기 초기화
words = okt.nouns(article_text)
tokens = okt.morphs(article_text)

print('토큰:',tokens)
print('단어:',words)

# 불용어 제거
stop_words = set(['은', '는', '이', '가', '을', '를', '의', '에', '와', '과', '한', '하다','것','약','위해','할', '수', '하고','라고','로','으로','할','수','보','주','등','의','-','.','"','"','(',')']) # 한글 불용어 사전 로드
filtered_tokens = [word for word in tokens if word not in stop_words]
filtered_words = [word for word in words if word not in stop_words]

# 결과 출력
print("불용어 제거 토큰:", filtered_tokens)
print("불용어 제거 단어:", filtered_words)
```

```python
import json

with open('SentiWord_info.json', encoding='utf-8-sig', mode='r') as f:
  SentiWord_info = json.load(f)

sentiword_dic = pd.DataFrame(SentiWord_info)
```

```python
import pandas as pd

df = pd.DataFrame(columns=("sentences", "sentiment"))  # 데이터프레임 생성
idx = 0                                             # 다음으로 넘기기 위한 초기값

for token in tokens:                                # 전체에서 문장 하나씩 가져옴
  sentiment = 0                                     # 초기 감성값 0으로 설정
  for i in range(0, len(sentiword_dic)):            # 감성사전의 모든 단어를 하나씩 선택
    if sentiword_dic.word[i] in token:              # 리뷰 문장에 감성 단어가 있는지 확인
      sentiment += int(sentiword_dic.polarity[i])   # 감성단어가 있다면 극성값 합계를 구함.
  df.loc[idx] = [token, sentiment]                  # 리뷰별 극성값을 데이터프레임으로 쌓음
  idx += 1                                          # 다름 리뷰 문장으로 넘어감
```
